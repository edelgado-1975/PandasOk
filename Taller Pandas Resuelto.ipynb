ADICIONAR UNA NUEVA COLUMNA

```python
# Su tarea es crear una columna llamada 'Critic Rating' que debe tener números enteros aleatorios entre 1 y 4.

# 1. Importa numpy y crea 550 números enteros aleatorios entre 1 y 4
import numpy as np
int_rating = np.random.randint(1, 4, 550)

# 2. Agregue la nueva columna 'Critic Rating' al marco de datos usando los números aleatorios creados
df_books['Critic Rating'] = int_rating

# Nota: Tenga en cuenta que los números aleatorios en esta nueva columna 'Critic Rating' serán diferentes entre su solución y la mía, pero nos centraremos solo en el código en esta sección.
```

Atributos basicos, Metodos y Funciones

```python
# obtener acceso al atributo de columnas
df_books.columns
```

Selecciona dos o mas columnas desde un Dataframe

```python
# Mueva la nueva columna 'Critic Rating' entre las columnas "Calificación del usuario" y "Reseñas". Luego actualice la dataframe

# Consejo: Copie y pegue los nombres de las columnas obtenidos con el atributo de columnas y luego reorganice los elementos usando [[]]
df_books = df_books[['Name', 'Author', 'User Rating', 'Critic Rating', 'Reviews', 'Price', 'Year', 'Genre']]
```

```python
# mostrar las primeras 5 filas
df_books.head()
```

Operaciones con Dataframes

```python
# Cree una columna denominada "Calificación promedio" utilizando la siguiente fórmula: Calificación promedio = (Calificación del usuario + Calificación del crítico)/2
df_books['Average Rating'] = (df_books['User Rating'] + df_books['Critic Rating'])/2
print(df_books)
```

```python
# Utilice la función round para redondear los valores del marco de datos a 1 decimal y actualizar el dataframe
df_books = round(df_books, 1)
#print(df_books)
```

Contar Valor

```python
# contar elementos en la columna "Género" por categoría y devolver la frecuencia relativa
df_books['Genre'].value_counts(normalize=True)
```

Renombrar Columnas

```python
# Cambie el nombre de las columnas "Calificación del usuario", "Calificación del crítico" y "Calificación promedio" a "UR", "CR" y "AR" y luego actualice el marco de datos con el parámetro incorporado.
df_books.rename(columns={'User Rating':'UR',
                         'Critic Rating':'CR',
                         'Average Rating':'AR'}, inplace=True)
```

```python
# mostrar las primeras 5 filas
df_books.head()
```

```python
# Seleccione solo las columnas "Nombre", "Autor", "UR", "CR", "AR" y "Año" y actualice el dataframe
df_books = df_books[["Name", "Author", "UR", "CR", "AR", 'Year']]
```

Ordenar un Dataframe

```python
# Ordena el dataframe en orden descendente por "UR" y "CR"
df_books.sort_values(['UR', 'CR'], ascending=False)
```
