CREANDO UN DATAFRAME DESDE UN ARREGLO EN PANDAS

```python
import pandas as pd
import numpy as np
```

OPCION 1

```python
# creamos el arreglo
data = np.array([[1, 4], [2, 5], [3, 6]])
```

```python
# creamos el dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],columns=['col1', 'col2'])
```

```python
# showing the datafrane
df
```

OPERACIONES CON DATAFRAMES

Operaciones Matematicas

```python
# count(contar), mean(media), std(desviacion standar), max(maximo), y minb (minimo)
df_examenes['math score'].count()
df_examenes['math score'].mean()
df_examenes['math score'].std()
df_examenes['math score'].max()
df_examenes['math score'].min()
```

```python
# Calculo mas sencillo usando .describe()
df_examenes.describe()
```

```python
# select a column and calculate total sum
df_examenes['math score'].sum()
```

Operaciones en filas

```python
# Calculando la suma en una fila
df_examenes['math score'] + df_examenes['reading score'] + df_examenes['writing score']
```

```python
# Calculando la puntuación media y asignando el resultado a una nueva columna
df_examenes['average'] = (df_examenes['math score'] + df_examenes['reading score'] + df_examenes['writing score'])/3
```

CONTAR VALORES EN EL DATAFRAME

```python
# contando elementos de género

# funcion len
len(df_examenes['gender'])
# metodo .count()
df_examenes['gender'].count()
```

```python
# Contando elementos de género por categoría
df_examenes['gender'].value_counts()
```

```python
# devuelve la frecuencia relativa (divide todos los valores por la suma de valores)
df_examenes['gender'].value_counts(normalize=True)
```

```python
# Conteo de elementos del "parental level of education" por categoría
df_examenes['parental level of education'].value_counts()
```

```python
# Devuelve la frecuencia relativa y la redondea a 2 decimales.
df_examenes['parental level of education'].value_counts(normalize=True).round(2)
```

ORDENAR UN DATAFRAME

```python
# ordenar por una columna
df_examenes.sort_values('math score')
```

```python
# ordenar descendentemente por una columna
df_examenes.sort_values('math score', ascending=False)
```

```python
# Ordenar descendentemente por múltiples columnas
df_examenes.sort_values(['math score', 'reading score'], ascending=False)
```

```python
#  Ordenar descendentemente por múltiples columnas y actualizar el dataframe
df_examenes.sort_values(['math score', 'reading score'], ascending=False, inplace=True)
```

```python
# Ordenar descendentemente con una función de tecla
df_examenes.sort_values(['race/ethnicity'], ascending=True, key=lambda a:a.str.lower())
```
